```python
# Para somar somente as colunas com números podemos excluir a primeira coluna 'id'
contagem_moa = dados_resultados.drop('id', axis=1).sum().sort_values(ascending = False)
contagem_moa
```

```python
# Outra maneira de somar os números nas colunas utilizando dtypes. MOA = mecanismo de ação.
contagem_moa = 0
contagem_moa = dados_resultados.select_dtypes('int64').sum().sort_values(ascending = False)
contagem_moa
```

```python
# Fazendo a soma das linhas.
dados_resultados.drop('id', axis=1).sum(axis=1)
```

```python
# Adicionando coluna com a soma das linhas na tabela de resultados. MOA = mecanismo de ação.
dados_resultados['n_moa'] = 0
dados_resultados['n_moa'] = dados_resultados.drop(['id','n_moa'], axis=1).sum(axis=1)
```

```python
dados_resultados['ativo_moa'] = (dados_resultados['n_moa']!=0)
dados_resultados.head()
```

```python
# Utilizando função merge para combinar os dados das duas tabelas
dados_combinados = pd.merge(dados, dados_resultados[['id','n_moa','ativo_moa']], on = 'id')
dados_combinados.head()
```

```python
# Verificando se as linhas com_controle realmente não causam mecanismo de ação
dados_combinados.query('tratamento == "com_controle"')['ativo_moa'].value_counts()
```

```python
# Verificando se todas drogas ativaram o mecanismo
dados_combinados.query('tratamento == "com_droga"')['ativo_moa'].value_counts()
```

```python
# Dado um composto -> obteve-se uma assinatura celular -> da assinatura obteve-se o mecanismo de ação que foi ativado (MOA).
# No machine learning -> vamos tentar ensinar à máquina deduzir qual MOA foi ativado baseada na assinhatura celular para cada composto.

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y,test_size = 0.2)

```

```python
from sklearn.linear_model import LogisticRegression


modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino,y_treino)
modelo_rlogistica.score(x_teste,y_teste)
```

```python
# Verificar o quão melhor o modelo logistica é melhor do que outros modelos
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino,y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste,previsao_dummy)
```

```python
# Nota-se que o modelo não apresenta melhora significativa com relação a outros modelos simplórios, como o "mais frequente"
# Os dois têm resultado próximo do percentual de True encontrado na base de dados analisada, conforme abaixo
dados_combinados['ativo_moa'].value_counts(normalize = True)
```

```python
# Alterando a forma como o modelo seleciona os dados para forma aleatória
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y,test_size = 0.2, stratify =y, random_state =376)
modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino,y_treino)
modelo_rlogistica.score(x_teste,y_teste)
```

```python
# Verificando do modelo mais frequente agora:
modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino,y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste,previsao_dummy)
```

```python
# Testando outro modelo:
from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y,test_size = 0.2, stratify =y, random_state =376)

modelo_arvore = DecisionTreeClassifier(max_depth = 3)
modelo_arvore.fit(x_treino,y_treino)
modelo_arvore.score(x_teste,y_teste)
```

```python
# Desenhando a árvore de decisão:

from sklearn import tree

fig, ax = plt.subplots(figsize = (15,10), facecolor = 'k')
tree.plot_tree(modelo_arvore, ax=ax, fontsize = 10, rounded = True, filled = True, feature_names = x_treino.columns, class_names = ['Não Ativado', 'Ativado'])
plt.show()
```

```python
# Verificando a acurácia do modelo de árvore de decisão para maiores níveis de profundidade, tanto para fase de teste como para a fase de treino.

from sklearn.tree import DecisionTreeClassifier

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y,test_size = 0.2, stratify =y, random_state =376)

teste = []
treino = []

for i in range(1,15):
  modelo_arvore = DecisionTreeClassifier(max_depth = i)
  modelo_arvore.fit(x_treino,y_treino)
  teste.append(modelo_arvore.score(x_teste,y_teste))
  treino.append(modelo_arvore.score(x_treino,y_treino))
```

```python
treino
```

```python
# Visualizando graficamente o efeito da melhoria de acurácia nos dados de treino e queda nos dados de teste.
# Problema de machine learning conhecido como Overfit.
sns.lineplot(x=range(1,15), y = teste, label = 'teste')
sns.lineplot(x=range(1,15), y = treino, label = 'treino')
```
